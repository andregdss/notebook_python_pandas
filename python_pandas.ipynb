# Pandas

### Conceitos:

- É uma biblioteca para **manipulação** e **análise** de dados;

- Carrega os dados em um objeto do tipo ``DataFrame``;

- Funções básicas:
    - filtrar;
    - ordenar;
    - agregar;
    - transformar;

### Funções:

#### 1) Funções de Importação:

##### 1.1) Importação/Leitura Geral

- `read_csv():` lê uma base de dados em formato .csv e converte em um dataframe;

- `read_excel():` lê uma base de dados em formato excel e converte em um dataframe;

- `read_json():` lê uma base de dados em formato .json e converte em um dataframe;

- `read_html():` lê uma base de dados em formato .html e converte em um dataframe;

- `read_sql():` lê uma base de dados em formato .sql e converte em um dataframe;

##### 1.2) Outras Importações: CSV

- `pd.read_csv(url, nrows= , usecols= ):` lê uma base de dados mas apenas nas linhas e/ou colunas especificadas;
    - `usecols= `: intervalo de colunas (pelo indice ou pelo nome);
    - `nrows= `: intervalo de linhas;
- `pd.read_csv(url, encoding= , sep= , skiprows= , skipfooter = , engine='python'):`
    - `enconding = `: indicar a codificação específica;
    - `sep= `: indicar qual separador o arquivo csv está usando (padrão é virgula);
    - `skiprows = `: pular N primeiras linhas do arquivo;
    - `skipfooter= `: pular N ultimas linhas do arquivo;
    - `engine='python'`: motor melhor para usar as puladas de linha;

##### 1.3) Outras Importações: EXCEL

- `'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True':`é preciso adicionar o **?raw=True** na url;

- `pd.read_excel(url, engine='openpyxl'):` e usar o parametro **engine='openpyxl'** dentro da função read_excel;

- `pd.read_excel(url, sheet_name= ):` para ler uma página (sheet) especifica do arquivo;

- `pd.read_excel(url, sheet_name= , usecols= , nrows= ):` para ler uma sheet especifica, e ainda colunas e linhas também especificadas;

- `pd.ExcelFile(url).sheet_names:` retorna o nome das páginas (sheets) presentes no arquivo excel;

##### 1.4) Outras Importações: JSON e HTML



JSON:

- `pd.json_normalize(df['coluna']):` normaliza a coluna aninhada do dataframe;

- `pd.json_normalize(data_frame, record_path='coluna_aninhada', meta=['colunas_mantidas']):` normaliza no caso de JSON com listas dentro de dicionários;


- `var_get = requests.get(url): ` requisição GET para pegar os dados da API;

- `var_loads = json.loads(var_get.text):` converte os dados de String JSON (na variavel do get) para objeto Python;

- `pd.DataFrame(var_loads):` converte de JSON (conteudo da variavel do loads) para DataFrame;

HTML:

- `pd.read_html(url):` gera uma lista com todas as tabelas (tag table) presentes na página web;
    - para acessar cada tabela: `data_frame[n]`;

#### 2) Funções de Informação / Consulta:

- `.info():` exibe as informações gerais (colunas, Non-Null, tipos) do DataFrame;

- `.head():` exibe as 5 primeiras linhas do DataFrame;
    - pode exibir N linhas => .head(n)

- `.tail():` exibe as 5 ultimas linhas do DataFrame;
    - pode exibir N linhas => .tail(n)

- `type():` retorna o tipo do objeto que foi colocado como parâmetro;

- `.unique():` retorna os valores únicos (sem repetição) presentes na coluna indicada (numa lista);

- `.nunique():` retorna a quantidade de elementos únicos da coluna indicada;

- `.isnull():` retorna um DataFrame booleano que indica (por linha), se tem dados nulos (True) ou não (False);
    - aplicando o `.sum()`, soma a quantidade de nulos;

- `data_frame['coluna'].value_counts:` retorna a quantidade de itens nessa coluna;
    - se usar o `.value_counts(normalize=True)` mostra porcetagens (ao inves de quantidades);

- `.shape:` retorna o número de linhas e colunas do DataFrame;

- `.columns:` retorna quais são as colunas do DataFrame (numa lista);

- `.dtypes:` retorna os tipos de dados de cada coluna do DataFrame (numa Series);

- `.index:` retona o índice da linha;



Consultas:

- `data_frame['coluna']:` exibe uma **Series** com índice e valor, para a coluna desejada;

- `data_frame[['coluna 1', 'coluna 2']]:` exibe um **DataFrame** com índice e valores para as duas (ou mais) colunas especificadas;

#### 3) Funções Estatísticas:

- `.describe():` retorna estatísticas básicas do DataFrame (por coluna);

- `.count():` retorna a quantidade de valores não nulos;

- `.mean():` retorna a média dos valores;

- `.std():` retorna o desvio padrão dos valores;

- `.min():` retorna o valor mínimo;

- `.max():` retorna o valor máximo;

- `.min():` retorna o valor mínimo;

- ``.quantile(0.25):`` retorna o 1º quartil ou 25% daqueles valores;

- ``.median():`` retorna a mediana daqueles valores;

- ``.quantile(0.75):`` retorna o 3º quartil ou 75% daqueles valores;

- `.size():` retorna o tamanho(quantidade) da coluna agrupada;

- `nlargest(N):` retorna os N maiores valores do agrupamento;

- `.rank(ascending= , method= ):` atribuir posições aos valores de uma Series ou coluna de um DataFrame;
    - `ascending=`: True (menor valor em 1) e False (maior valor em 1);
    - `method=`:
        - `average:` média das posições empatadas;
        - `min:` menor posição do empate;
        - `max:` maior posição do empate;
        - `first:` ordem de aparição;
        - `dense:` sequência contínua (não pula números);

- `data_frame.sum():` por padrão `axis=0`, soma verticalmente, ou seja, soma cada coluna inteira;

- `data_frame.sum(axis=1):` soma horizontalmente, ou seja, soma os valores de todas as colunas em cada linha;




#### 4) Funções de Ajustes Gerais:

##### 4.1) Criação e Estrutura de DataFrames / Series

- `pd.DataFrame(dados, index= , columns= ):` cria um DataFrame;
    - `dados`:
        - dicionário de listas: cada chave vira uma coluna, e os valores em cada linha;
        - lista de listas: cada lista interna é uma linha;
        - lista simples: cria uma coluna e cada elemento da lista fica numa linha;
        - dicionario de dicionarios: as chaves externas são colunas e chaves internas são indices;
        - series: alinha automaticamente pelos indices;
    - `index=`: define os rótulos das linhas;
    - `columns=`: define os nomes das colunas;

- `.to_frame():` transforma uma Series em um DataFrame;

- `.to_numpy():` converte o DataFrame em um array NumPy;

##### 4.2) Índices e Estrutura do DataFrame

- `.index.name = 'nome':` define um nome para o índice do DataFrame;

- `.reset_index():` usado para tirar um índice (com dados) e transformá-lo em coluna comum, voltando a usar um índice numérico padrão (0, 1, 2...);

- `.set_index('coluna', inplace=True):` define qual coluna será usada como índice do DataFrame;
    - `inplace=True`: modifica o próprio DataFrame, em vez de criar uma cópia;

- `.reindex():` é usado para reorganizar, adicionar ou remover índices (linhas ou colunas) de um DataFrame, alinhando-o a um novo conjunto de rótulos;

- `.unstack():` mover um nível do índice (geralmente um MultiIndex) para colunas, “desempilhando” os dados;

##### 4.3) Colunas e Estrutura do DataFrame

- `.rename_axis(index={'nome': 'nome_novo'}, columns='nome_novo')`: renomeia rotulo dos indices e nome das colunas;

- `.rename(columns={'nome':'nome_novo'}, inplace=True):` renomeia a coluna especificada, o inplace=True serve para que a alteração seja definitivamente aplicada;

- `.assign(coluna= ):` adiciona uma nova coluna no DataFrame, criando uma cópia dele;

- `data_frame.insert(n, 'coluna', valores):` adiciona uma nova coluna (e valores dela) num DataFrame ou Series:
    - `n:` posição onde a nova coluna será inserida;
    - `'coluna':` nome da nova coluna;
    - `valores:` valores que preencherão essa coluna (lista, Series ou array com o mesmo tamanho do DataFrame);

- `.drop():` remove *linhas (axis = 0)* ou *colunas (axis = 1)*, especificando os indices como parametro;

##### 4.4) Tipos de Dados e Datas

- `data_frame['colunas'].astype(np.int64 ou np.float64):` converte o dtype dos dados da coluna referida para o tipo especificado;

- `pd.to_datetime(data_frame['coluna']:)` converte o dtype dos dados da coluna para datetime64[ns];

- `data_frame['coluna'].dt.strftime('%Y-%m'):` formata a data para ano-mês;

- `.dt.days`: pega só o número inteiro de dias de uma diferença entre datas;

##### 4.5) Tratamento de Valores Nulos

- `.fillna():` substitui os valores nulos pelo que estiver passado como parametro;

- `.dropna():` remove todas as linhas/colunas que possuem pelo menos um valor nulo;

- `.interpolate():` interpola valores nulos usando os valores vizinhos;

##### 4.6) Operações com Strings

- `.replace('texto', 'novo texto'):` substitui partes de uma string por outra;
    - `regex=True:` `r'[^a-zA-Z0-9\-\']'`: abarca tudo que NÃO(`^`) for letra, número, hífen ou apóstrofo;
    - `regex=True:` `r'(?<!\w)-(?!\w)'`: abarca hífens que estão soltos, isolados, ou cercados por espaços;
    - `regex=True:` `r'\{|}|\"'`: abarca chaves e aspas duplas da string;

- `.strip():` remove espaços em branco do início e do fim da string;

##### 4.7) Seleção, Filtragem e Consulta de Dados

- `data_frame.loc['nome_indice']:` cria uma nova linha (indice);

- `data_frame.iloc[intervalo]:` seleciona dados pela posição numérica (índices inteiros) indicada no intervalo;

- `data_frame = data_frame.loc[filtro, colunas]:` usado para filtrar linhas e colunas ao mesmo tempo, usando rótulos (filtro);

- `data_frame = data_frame[filtro]:` usado para filtrar linhas especificas (exibe todas as colunas);

- `.query('comando sql'):` usada para filtrar dados (gera apenas uma cópia filtrada para visualização/consulta, é necessário atribuir);

- `data_frame['coluna'].str.contains(regex, na=False):` retorna apenas as linhas que contem o regex indicado;
    - `na=False:` parametro opcional que faz ignorar os valores nulos;

##### 4.8) Ordenação, Agrupamento e Agregação

- `.sort_values('coluna', ascending= ):` ordena os valores do DataFrame/Series pela coluna indicada;
    - `ascending = True`: ordem crescente;
    - `ascending = False`: ordem descrescente;


- `data_frame['coluna'].value_counts():` contagem de quantas vezes cada valor aparece na coluna especificada (retorna uma series);
    - `normalize=True`: retorna a proporção (frequência relativa) de cada valor;

- `.groupby('coluna 1')['coluna 2'].func_est():` gera uma Series ou Dataframe que agrupa os dados pela *coluna 1* (indices) e usa os valores da *coluna 2* (coluna), e aplica uma *função estatistica*;
    - `['coluna 2']` → Series
    - `[['coluna 2']]` → DataFrame

- `.groupby(['coluna 1', 'coluna 2'])[['coluna 2']].func_est():` gera um DataFrame com Multi-Index, ou seja, são feito dois agrupamentos hierarquicos;
    - `.xs('valor', level= ):` consulta dentro do DataFrame Multi-Index onde é especificado o level (qual das colunas indices) e o valor dentro dessa coluna;

- `.groupby('coluna').get_group('valor')` retorna o subconjunto do DataFrame correspondente ao grupo em que a coluna possui o valor especificado;


##### 4.9) Junção de DataFrames (merge)

- `pd.merge(data_frame 1, data_frame 2, left_on='coluna A', right_on='coluna B'):` junta dois DataFrames, organizando as linhas por valor da coluna A (esq) sendo igual ao da coluna B (dir);

- `pd.merge(data_frame 1, data_frame 2, on=, how= ):` junta dois dataframes;
    - `on=`: nome da coluna em comum;
    - `how=`: forma da junção;
        - `'inner'`: apenas os registros que existirem nos dois DataFrames serão mantidos;
        - `'left'`: mantém todos os do primeiro DataFrame, e coloca NaN onde não há correspondência;
        - `'right'`: mantém todos os do segundo DataFrame;
        - `'outer'`: mantém todos de ambos, preenchendo NaN onde não houver correspondência

##### 4.10) Reorganização e Reshape dos Dados

- `.melt(id_vars= , value_vars= , var_name= , value_name= ):` reorganiza um DataFrame do formato largo para o formato longo, transformando colunas em linhas:
    - `id_vars:` colunas que permanecem fixas e identificam cada registro;
    - `value_vars:` colunas que serão transformadas em linhas;
    - `var_name:` nome da nova coluna que armazenará os nomes das colunas transformadas;
    - `value_name:` nome da nova coluna que armazenará os valores dessas colunas;

- `data_frame.pivot_table(index= , columns= , values= ):` reorganiza os dados criando uma tabela resumida;
    - `index:` indica a coluna que os valores irão virar indices da tabela;
    - `columns:` indica a coluna do DataFrame que os valores irão virar colunas da tabela;
    - `values:` indica a coluna que os valores irão virar os valores da tabela;
    - `aggfunc`: aplica uma função de agregação nos valores;
    - `sort`: decide se ordena ou não os dados;

- `.explode():` se uma coluna tem listas, essa função cria uma linha por item da lista, repetindo o restante dos valores da linha;

##### 4.11) Aplicação de Funções

- `.apply(lambda function):` aplica uma função a cada elemento de uma Series ou a cada linha/coluna de um DataFrame;

- `.applymap(lambda function):` usada em DataFrames inteiros, aplicando a função em cada célula individual;

#### 5) Funções de Salvar/Exportar:

- `.to_csv('nome_arquivo.csv', index=False):` salva o DataFrame em um arquivo no formato .csv;

- `.to_json('nome_arquivo.json', index=False):` salva o DataFrame em um arquivo no formato .json;

- `.to_excel('nome_arquivo', index=False):` salva o DataFrame em um arquivo no formato excel;

- `.to_sql('nome_arquivo', index=False):` salva o DataFrame em um arquivo no formato sql;


#### 6) Funções de Banco de Dados:

- `engine = create_engine('sqlite:///meu_banco.db'):` cria o banco de dados local que vai ser alimentado;

- `data_frame.to_sql('nome_tabela', engine, index=False):` cria uma tabela no banco de dados com o DataFrame especificado;

- `data_frame.to_sql('nova_tabela', con=engine, index=False):` pega um DataFrame e adiciona como uma nova tabela ao banco de dados;

- `inspector = inspect(engine) / inspector.get_table_names():` retorna quais são as tabelas presentes no banco de dados;

- `query = ' ' / data_frame = pd.read_sql(query, engine):` faz uma query (filtro/visualização), e o resultado dela é salvo em uma DataFrame;

- `data_frame = pd.read_sql_table('nome_tabela', engine):` lê uma tabela dentro do banco de dados, e salva em um DataFrame;

- `pd.read_sql_table('nome_tabela', engine, columns=['colunas]):` lê apenas colunas específicas de uma determinada tabela;

- Fazendo uma alteração diretamente no DataFrame:


#### 7) Funções de Gráfico:

- `.plot(kind= , figsize= , color= ):` gera um gráfico de acordo com os parametros especificados;
    - `kind:` tipo do gráfico;
    - `figsize:` tamanho da figura onde o gráfico será plotado;

### Exercício 3: Base de Dados - Vendas e Clientes

Importando os dados:

````python
import pandas as pd
import numpy as np
````

````python
url_vendas = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'
df_vendas = pd.read_json(url_vendas)
df_vendas
````

Normalizando o DataFrame:

````python
df_vendas = pd.json_normalize(df_vendas['dados_vendas'])
df_vendas
````

"Explodindo" as colunas com listas:

````python
col_vendas = list(df_vendas.columns)

df_vendas = df_vendas.explode(col_vendas[1:])

df_vendas.reset_index(inplace=True, drop=True)
````

````python
df_vendas.head()
````

Ajustando a formatação dos números na coluna 'Valor da compra':

````python
df_vendas['Valor da compra'] = df_vendas['Valor da compra'].apply(lambda x: x.replace('R$', '').replace(',', '.').strip())
df_vendas['Valor da compra'] = df_vendas['Valor da compra'].astype(np.float64)
````

````python
df_vendas.info()
````

Ajustando a formatação da coluna 'Cliente':

````python
df_vendas['Cliente'] = df_vendas['Cliente'].str.lower()
df_vendas['Cliente'] = df_vendas['Cliente'].str.replace(r'[^A-Za-z]', '', regex=True)
df_vendas.head()
````

Ajustando os dados da coluna 'Data de venda':

````python
df_vendas['Data de venda'] = pd.to_datetime(df_vendas['Data de venda'], format='%d/%m/%Y')
df_vendas.info()
````

Criando agrupamento de Cliente pelo somatório do Valor de Compra:

````python
df_compras = df_vendas.groupby(['Cliente'])['Valor da compra'].sum().sort_values(ascending=False).to_frame()
df_compras
````

````python
df_compras = df_compras.reset_index()
df_compras
````

Salvando os DataFrames gerados:

````python
df_vendas.to_csv('dados_vendas.csv', index=False)
df_compras.to_csv('dados_compras.csv', index=False)
````

### Exercício 4: Base de Dados -  Locação de Imóveis

Importando os dados:

````python
url_locacao = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_locacao_imoveis.json'
df_locacao = pd.read_json(url_locacao)
df_locacao
````

Normalizando o DataFrame:

````python
df_locacao = pd.json_normalize(df_locacao['dados_locacao'])
df_locacao
````

"Explodindo" as colunas com listas:

````python
col_locacao = list(df_locacao.columns)

df_locacao = df_locacao.explode(col_locacao[1:])

df_locacao.reset_index(inplace=True, drop=True)
````

Ajustando a formatação dos números na coluna 'valor aluguel':

````python
df_locacao['valor_aluguel'] = df_locacao['valor_aluguel'].apply(lambda x: x.replace('$', '').replace(',', '.').replace('reais', '').strip())
df_locacao['valor_aluguel'] = df_locacao['valor_aluguel'].astype(np.float64)
````

Ajustando a formatação da coluna 'apartamento':

````python
df_locacao['apartamento'] = df_locacao['apartamento'].str.replace('(blocoAP)', '').str.strip()
df_locacao.head()
````

Ajustando o tipo e formatação das colunas de datas:

````python
df_locacao['datas_combinadas_pagamento'] = pd.to_datetime(df_locacao['datas_combinadas_pagamento'], format='%d/%m/%Y')
df_locacao['datas_de_pagamento'] = pd.to_datetime(df_locacao['datas_de_pagamento'], format='%d/%m/%Y')
df_locacao.info()
````

Criando a coluna de dias de atraso:

````python
df_locacao['atraso'] = (df_locacao['datas_de_pagamento'] - df_locacao['datas_combinadas_pagamento']).dt.days
df_locacao
````

Criando um Dataframe de agrupamento por Apartamento e soma dos dias de Atraso:

````python
df_locacao_atraso = df_locacao.groupby(['apartamento'])['atraso'].sum().to_frame().reset_index()
````
